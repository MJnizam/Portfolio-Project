# Salary Data Analysis

This analysis explores the relationship between Years of Experience and Salary.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Salary_dataset.csv')

# Remove unnecessary index column
df_cleaned = df.drop(columns=['Unnamed: 0'])

# Display basic info
df_cleaned.info(), df_cleaned.head()
```

## Data Cleaning
- Removed an unnecessary index column.
- Checked for duplicates (None found).
- Summary statistics below:

```python
df_cleaned.describe()
```

## Data Visualization
### 1. Salary Distribution
### 2. Salary vs. Years of Experience

```python
sns.set_style('whitegrid')

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Salary Distribution
sns.histplot(df_cleaned['Salary'], bins=10, kde=True, ax=axes[0], color='royalblue')
axes[0].set_title('Salary Distribution')
axes[0].set_xlabel('Salary')
axes[0].set_ylabel('Frequency')

# Salary vs. Years of Experience
sns.scatterplot(x=df_cleaned['YearsExperience'], y=df_cleaned['Salary'], ax=axes[1], color='darkorange')
axes[1].set_title('Salary vs. Years of Experience')
axes[1].set_xlabel('Years of Experience')
axes[1].set_ylabel('Salary')

plt.tight_layout()
plt.show()
```

## Key Findings
- Salary increases with experience.
- Data distribution shows most salaries between 40,000 and 100,000.
- No missing values or duplicates were found.
